```python
import pandas as pd
```

```python
iris.feature_names
```

```python
t = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
ta = pd.DataFrame(iris.target, columns=['target'])
```

```python
data = pd.concat([t,ta], axis=1)
```

```python
data
```

```python
data.info()
```

---

```python
from sklearn.model_selection import learning_curve
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
data[data.columns[:-1]]
```

```python
train_size, train_scores, test_scores = \
    learning_curve(KNeighborsClassifier(), data[data.columns[:-1]], data.iloc[:,-1], cv=10)
```

```python
train_size
```

```python
train_scores
```

```python
test_scores
```

```python
from sklearn.datasets import make_classification
```

```python
data = make_classification(n_samples=1000, n_features=5)
```

```python
data[0]
```

```python
xdata = pd.DataFrame(data[0])
```

```python
tdata = pd.DataFrame(data[1],columns=['target'])
```

```python
testdata = pd.concat([xdata,tdata],axis=1)
```

```python
testdata
```

```python
testdata.info()
```

```python
#Lo
train_size, train_scores, test_scores = \
    learning_curve(LogisticRegression(), testdata.iloc[:,:-1],testdata.iloc[:,-1], cv=10)
```

```python
plot.learning_curve(train_scores, test_scores, train_size)
```

```python
#KN
%time
train_size, train_scores, test_scores = \
    learning_curve(KNeighborsClassifier(), testdata.iloc[:,:-1],testdata.iloc[:,-1], cv=10)
```

```python
plot.learning_curve(train_scores, test_scores, train_size)
```

- KN알고리즘은 데이터의 양이 많아진다고 성능이 좋아지지 않는다.(instance base, case)

```python
from sklearn.svm import SVC
```

```python
train_size, train_scores, test_scores = \
    learning_curve(SVC(), testdata.iloc[:,:-1],testdata.iloc[:,-1], cv=10)
```

```python
plot.learning_curve(train_scores, test_scores, train_size)
```

## No Free Lunch
    - 데이터에 따라서 최적의 알고리즘이 있고, 알고리즘에 따라 필요한 데이터양이 따로 있다.
    - 따라서 데이터가 적을 때에는 적은 데이터에서 잘 작동하는 알고리즘을 선택한다.
