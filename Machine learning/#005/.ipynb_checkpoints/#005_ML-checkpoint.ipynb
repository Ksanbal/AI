```python
from sklearn.datasets import make_classification
```

```python
data = make_classification(n_samples=1000, n_features=5)
```

```python
data[0]
```

```python
xdata = pd.DataFrame(data[0])
```

```python
tdata = pd.DataFrame(data[1],columns=['target'])
```

```python
testdata = pd.concat([xdata,tdata],axis=1)
```

```python
testdata
```

```python
testdata.info()
```

```python
#Lo
train_size, train_scores, test_scores = \
    learning_curve(LogisticRegression(), testdata.iloc[:,:-1],testdata.iloc[:,-1], cv=10)
```

```python
plot.learning_curve(train_scores, test_scores, train_size)
```

```python
#KN
%time
train_size, train_scores, test_scores = \
    learning_curve(KNeighborsClassifier(), testdata.iloc[:,:-1],testdata.iloc[:,-1], cv=10)
```

```python
plot.learning_curve(train_scores, test_scores, train_size)
```

- KN알고리즘은 데이터의 양이 많아진다고 성능이 좋아지지 않는다.(instance base, case)

```python
from sklearn.svm import SVC
```

```python
train_size, train_scores, test_scores = \
    learning_curve(SVC(), testdata.iloc[:,:-1],testdata.iloc[:,-1], cv=10)
```

```python
plot.learning_curve(train_scores, test_scores, train_size)
```

### 하이퍼 파라미터 찾기

```python
from sklearn.model_selection import train_test_split
```

```python
iris
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1], data.iloc[:,-1])
```

```python
temp = []
for i in range(2,21):
    knn = KNeighborsClassifier(i)
    knn.fit(X_train, y_train)
    temp.append(knn.score(X_test,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
para_grid = {'n_neighbors':range(2,21), 'weights':['uniform','distance']}
```

```python
gri = GridSearchCV(KNeighborsClassifier(), para_grid, cv=10, n_jobs=-1)
```

```python
gri.fit(data.iloc[:,:-1],data.iloc[:,-1])
```
