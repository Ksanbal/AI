```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(8)
```

```python
vars(pca)
```

```python
t = pca.fit_transform(wine.iloc[:,:-1])
```

```python
x = pd.DataFrame(t)
```

```python
x
```

```python
wine_pca = pd.concat([x,wine.target], axis=1)
```

```python
wine_pca
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
#wine ë°ì´í„°
cross_val_score(KNeighborsClassifier(),wine.iloc[:,:-1],wine.target,cv=10).mean()
```

```python
#wine_pca ë°ì´í„°
cross_val_score(KNeighborsClassifier(),wine_pca.iloc[:,:-1],wine_pca.target,cv=10).mean()
```
