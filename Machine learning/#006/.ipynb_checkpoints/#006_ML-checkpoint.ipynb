## 차원축소
    - 2차원 데이터를 1차원으로 옮겨, 구분할수있는 특징은 남긴다.
    - 차원이 커지면 필요한 데이터양도 많아야 한다. -> 차원축소를 통해 필요한 데이터양을 줄일 수 있다.
    - column의 값을 조정하기에 target도 조정해야하지만 target의 변경은 문제가 있다.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(8)
```

```python
vars(pca)
```

```python
t = pca.fit_transform(wine.iloc[:,:-1])
```

```python
x = pd.DataFrame(t)
```

```python
x
```

```python
wine_pca = pd.concat([x,wine.target], axis=1)
```

```python
wine_pca
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
#wine 데이터
cross_val_score(KNeighborsClassifier(),wine.iloc[:,:-1],wine.target,cv=10).mean()
```

```python
#wine_pca 데이터
cross_val_score(KNeighborsClassifier(),wine_pca.iloc[:,:-1],wine_pca.target,cv=10).mean()
```
