```python
tips.iloc[:5,:3] #파이썬의 index처럼 사용하는 방법
```

```python
tips.iat[0,5] #무조건 하나의 값만 가져온다.
```

```python
x = tips.groupby('sex').mean()#문자 카테고리 기준으로 묶을 수 있다.
x
```

```python
x['ratio'] = x.tip/x.total_bill
x
```

```python
t = tips.groupby('smoker').mean()
t['ratio'] = t.tip/t.total_bill
t
```

```python
tips.groupby(['sex','day']).mean()
```

```python
tips.groupby(['sex','day']).max()
```
