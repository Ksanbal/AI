### 시간을 기준으로 표 그리기 -> 시계열 분석

```python
dust = pd.read_csv('v.csv',engine = 'python', encoding = 'cp949')
dust
```

```python
dust.set_index('날짜')[dust.columns[1]] # index를 지정한다.
```

```python
import matplotlib as plot
```

```python
dust.set_index('날짜')[dust.columns[1]].plot.line()
```

```python
dust.info()
```

```python
dust['date'] = pd.to_datetime(dust['날짜'])
```

```python
dust
```

```python
dust.set_index('date')[dust.columns[1]].resample('m').mean().plot.line() # 월별로 그린 그래프
```

```python
dust.set_index('date')[dust.columns[1]].plot.line() #일별로 그린 그래프
```
