```python
import pandas as pd
```

```python
data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header = None)
# 웹에서도 데이터를 동적으로 가져올 수 있다.
```

```python
data # Column name이 없을 경우 첫 행의 데이터를 Column name으로 사용한다. 그 경우 header를 None으로 해줘야한다.
```

```python
data.info()
```

```python
data.describe()
```

```python
data.boxplot() #describe대신 그래프로 정보를 볼 수 있다.
```

```python
pd.plotting.scatter_matrix(data) # 정규분포인지 아닌지를 알게해준다.
```

```python
import seaborn as sns #pandas를 기반으로 통계적 그래프를 그려준다.
```

```python
sns.pairplot(data)
```

```python
data.rename({0:'sl', 1:'sw', 2:'pl', 3:'pw', 4:'class_'}, axis=1, inplace=True) #inplace를 True로 두면 원래 식별자가 변경된다.
# class라는 이름을 사용할때는 파이썬과 중복 문제가 있어서 관례상 뒤에 _를 붙여준다.
```

```python
data
```

```python
data.class #파이썬의 class와 중복되는 문제
```

```python
sns.pairplot(data, hue='class_') #hue : 선택한 colum name을 기준으로 색을 나눠준다.
```
