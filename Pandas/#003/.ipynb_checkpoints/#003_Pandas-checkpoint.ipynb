### Where()

- 

```python
a = np.arange(0,10)
a
```

```python
np.where(a>4,1,0) #뒤에 조건이 붙으면 값을 변경시킬 수 있다.
```

```python
np.where(a==4) # numpy에는 index번호를 불러올 수 없기 때문에 사용된다.
```

```python
billboard2.where() #조건에 맞으면 변화x, 조건이 맞지않으면 변화O 
```
