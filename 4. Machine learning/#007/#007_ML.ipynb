## - unsupervised learning

- 클러스터링 : 임의로 묶음(하이퍼 파라미터)이 존재한다고 가정한 후 주변 값들의 평균을 계속 찾아가면서 분류한다.(거리기반)

- 단점 : 하이퍼 파라미터를 정해줘야한다, 랜덤으로 점을 찍기에 특정데이터에서 분류오류가 난다, 섞여있는 데이터를 잘 구분하지 못한다.
