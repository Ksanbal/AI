## - unsupervised learning

- 클러스터링 : 임의로 묶음(하이퍼 파라미터)이 존재한다고 가정한 후 주변 값들의 평균을 계속 찾아가면서 분류한다.(거리기반)
- 분류된 값을 가지고 타겟데이터로 만들어 supervise 기법을 이용할 수 있다.
- 단점 : 하이퍼 파라미터를 정해줘야한다, 랜덤으로 점을 찍기에 특정데이터에서 분류오류가 난다, 섞여있는 데이터를 잘 구분하지 못한다.

```python
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
data = load_iris()
```

```python
iris = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
sns.pairplot(iris)
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(3)
```

```python
km.fit(iris.values) #.values = numpy값으로 변환
```

```python
vars(km)
```

```python
iris['target'] = km.labels_
```
