```python
#index가 유저, movie가 컬럼, rating을 값으로 변환
rating = data.set_index(['user_id','movie_id']).unstack().fillna(0)
rating
```

```python
#상관관계는 컬럼끼리 계산하기 때문에 행과열을 바꾼다.
rating_corr = rating.T.corr()
```
