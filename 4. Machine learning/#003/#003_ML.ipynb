import seaborn as sns
import pandas as pd

```python
data = sns.load_dataset('iris')
```

```python
#pandas_one hot
pd.get_dummies(data.species)
```

```python
#사이킷_one hot
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit_transform(data[['species']]).toarray() #인자로 데이터 프레임이 들어가야 한다. 바꾼값에 0이 많으면 특수한 형태로 저장되기 때문에 .toarray()로 봐야한다.
# ohe.fit(data[['species']]) 
# ohe.transform(data[['species']]).toarray()
```

```python
ohe.inverse_transform([[1,0,0]]) #내가 바꾼 값이 원래 뭐였는지 알 수 있다.
```

```python
#사이킷_label
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit_transform(data.species)
```

```python
x = pd.DataFrame(le.fit_transform(data.species), columns=['target']) #pandas형태로 변환
x
```

```python
data_pd = pd.concat([data,x], axis=1)
data_pd
```

## 성능평가
    - 알고리즘, 하이퍼 파라미터에 따라 성능차이가 난다.

### Holdout
    - data leakage : 데이터가 작을 경우 성능의 차이가 크게 난다.
    - test와 train을 나눈 후, train 데이터를 한번 더 나눠서 최종 성능을 처음에 뽑은 test로 한다.

```python
from sklearn.model_selection import train_test_split
```

```python
#colums로 자르기
X_train,X_test,y_train,y_test = train_test_split(data_pd[data_pd.columns[:-2]], data_pd.target)
```

```python
#.iloc로 자르기
X_train,X_test,y_train,y_test = train_test_split(data_pd.iloc[:,:-2], data_pd.iloc[:,-1])
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn5 = KNeighborsClassifier(5)
knn7 = KNeighborsClassifier(7)
```

```python
knn5.fit(X_train,y_train)
knn7.fit(X_train,y_train)
```

```python
knn5.score(X_test,y_test) #모델의 정확도를 측정해준다.
```

```python
knn7.score(X_test,y_test)
```

### Cross-validation
    - 데이터를 k-fold(k등분)하여 train과 test를 나눠 10번에 거쳐 학습시킨다.
    - 실무에서는 10-fold가 주로이다.
    - 단점 : 데이터가 커지면 처리시간이 커진다.
    - 사용목적 : 대충의 성능파악, 하이퍼 파라미터 찾기

```python
from sklearn.model_selection import cross_val_score
```

```python
t = KNeighborsClassifier()
k = cross_val_score(t, data_pd[data_pd.columns[:-2]],data_pd.target, cv=10)
```

```python
import numpy as np
```

```python
np.mean(k)
```

```python
for i,j in enumerate(range(1,21)):
    t = KNeighborsClassifier(j+1)
    k = cross_val_score(t, data_pd[data_pd.columns[:-2]],data_pd.target, cv=10)
    print(j,np.mean(k))
```
