- 맛보기
- sklearn
    - load_, fetch_, make_ -> 동사_ : function
    - load_ : 데이터 사이즈가 작아서 탑재되어있음
    - fetch_ : 데이터 사이즈가 커서 웹에서 불러온다.
    - make_ : 랜덤하게 데이터를 만들어준다.
- 방법(알고리즘)과 우리가 지정한(하이퍼 파라미터)를 이용해 컴퓨터가 판단한다.

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
type(data) #Bunch : 딕셔너리형태, 어트리뷰트 방식 접근
```

```python
print(data.DESCR) #데이터로 보여지는 정보외의 외부적인 정보들
```

```python
data.data, data.target # 데이터, 정답
```

```python
import pandas as pd
```

```python
data_pd = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
data_pd
```

```python
data_target = pd.DataFrame(data.target, columns=['target']) #관례상 target,class,label이라 이름 붙임
data_target
```

```python
#두개의 데이터를 하나의 데이터로 합친다.
iris = pd.concat([data_pd, data_target], axis=1)
iris
```

```python
iris.info()
```

```python
iris.describe()
```

```python
import seaborn as sns
```

```python
sns.pairplot(iris, hue='target', vars=iris.columns[:-1]) #vars를 통해서 숫자컬럼만 남기고, 타겟으로 분류한다.
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(iris.iloc[:,:-1], iris.iloc[:,-1])
```

```python
knn.predict([[3,3,4,3]])
```

```python
knn.predict_proba([[3,3,4,3]])
```
