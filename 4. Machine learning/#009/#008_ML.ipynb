## Validation curve
    - grid search CV에서의 성능만으로 하이퍼 파라미터를 선정할 수 없기 때문에, 그래프를 그려서 알맞은 파라미터를 찾기 위해서 사용한다.

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.species = iris.species.map({'setosa':0,'versicolor':1,'virginica':2})
```

```python
iris
```

```python
from sklearn.model_selection import learning_curve, validation_curve
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
```

```python
train_score, test_score = validation_curve(knn, iris.iloc[:,:-1], iris.iloc[:,-1], 'n_neighbors', [3,4,5,6,7], cv=10)
```

```python
from sklearn_evaluation import plot
```

```python
plot.validation_curve(train_score, test_score, [3,4,5,6,7], 'n_neighbors')
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
import numpy as np
```

```python
# Load data
data = sm.datasets.get_rdataset("Guerry", "HistData").data
```

```python
data
```

```python
dir(data)
```

```python
# Fit regression model (using the natural log of one of the regressors)
ols = smf.ols('Lottery ~ Literacy + Instruction', data=data)
# (y = ax + b) -> formula
ols = ols.fit()
```

```python
# Inspect the results
ols.summary()
```
