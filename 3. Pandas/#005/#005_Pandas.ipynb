### Matplotlib

- 종류 : pyplot, pylab(무겁고 충돌이 많아서 권고되지 않음)

```python
import matplotlib.pyplot as plt
```

```python
plt.figure() #눈에 보이지 않는 가상의 영역 -> 별도 선언하지 않아도, 하위 객체 생성시 자동으로 생성된다.
```

```python
plt.style.available #사용할 수 있는 스타일 종류
```

```python
plt.style.use('ggplot') #표의 스타일을 변경할 수 있다
```

```python
plt.plot([1,2,3],[4,5,6])
```

```python
plt.figure(figsize=(10,5), facecolor='pink') #figure와 axes는 순서가 중요하다.
plt.axes(ylim=(0,15)) #ylim : y축의 limit를 건다.
plt.title('Title') #한글은 따로 처리하지 않으면, 깨져서 보인다.
plt.xlabel('Xlabel') #x축 라벨
plt.ylabel('Ylabel') #y축 라벨
plt.grid(True, axis='y') #표 내부에 그리드를 그려준다.
plt.axis([1,10,1,10])
plt.plot([1,2,3],[4,5,6], color='k', marker='o', label='ttt') #라인 그래프를 그려준다. 가변키워드이지만 특정 옵션이 아닌 것을 넣으면 에러가 난다.
plt.legend(loc=2) #loc로 legend의 위치를 바꿀수 있다.
```

- 표안에 표를 넣을 수도 있다.

```python
plt.axes([0,0,1,1])#간격, 사이즈[left, right, width, hight]
plt.axes([0.3,0.3,0.5,0.5])
```

```python
plt.bar([1,2,3],[4,5,6])
```

### Seaborn 

- sns.*에서 plot이 붙어있는 함수는 그래프 함수
- pandas를 데이터로 사용한다.
- data 자리에 데이터폼을 넣는다.

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris, hue='species') #hue:지정한 컬럼이름을 기준으로 색을 다르게 하겠다.
# plt.axes(facecolor='pink')
```
