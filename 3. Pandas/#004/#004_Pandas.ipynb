```python
tips.pivot_table(index='smoker',columns='sex', aggfunc=np.sum, margins=True) #기본값은 평균, margins=여백자리에 all
```

```python
pd.crosstab([tips.smoker, tips.sex], tips.time, values=tips.tip, aggfunc=np.mean) #aggfunc=count가 기본, values는 안에 들어간 값
#계층적 index
```

```python
x = pd.crosstab([tips.smoker, tips.sex], tips.time, values=tips.tip, aggfunc=np.mean).index
x
```

- levels : index쪽에서 나누는 것

```python
x.codes #append는 가능하나, 모양을 똑같이 맞춰야해서 FrozenList라 지어졌다.
```

```python
x['levels'] #구조의 형태가 =으로로 이뤄진 구조는 Attribute 방식으로 접근해야한다.
```

### MultiIndex graph

- 무조건 index는 x축
- index를 묶어서 x축으로 보낸다.

```python
tips.groupby('sex').mean()[['tip']].unstack()
```

```python
tips.groupby('sex').mean()[['tip']].unstack().plot.bar()
```

```python
tips.groupby('day').mean()[['tip']].unstack().plot.bar(stacked=False)
```

```python
tips.groupby(['day','sex']).mean()[['tip']].plot.bar()
```

```python
tips.groupby(['day','sex']).mean()[['tip']].unstack()
```

```python
tips.groupby(['day','sex']).mean()[['tip']].unstack().plot.bar() #Column마다 색을 다르게 칠해준다.
```

```python
tips.groupby(['day','sex']).mean()[['tip']].unstack().plot.bar(stacked=True)
```

```python
tips.groupby(['day','sex']).mean()[['tip','total_bill']].plot.bar()
```

```python
tips.groupby(['day','sex']).mean()[['tip','total_bill']].unstack().plot.bar()
```

```python
tips.groupby(['day','sex']).mean()[['tip','total_bill']].unstack().plot.bar(stacked=True)
```

- 컬럼으로 보낼 인덱스 고르기
    - unstack(0)과 unstack()

```python
tips.groupby(['day','sex']).mean()[['tip','total_bill']].unstack(0).plot.bar(stacked=True)
```
