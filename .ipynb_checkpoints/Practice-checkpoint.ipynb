```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
data.columns[:-1]
```

```python
data.iloc[:,:-1]
```

```python
knn.fit(data.iloc[:,:-1], data.species)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.iloc[:,:-1],data.species)
```

```python
knn.fit(X_train,y_train)
```

```python
knn.predict(X_test) == y_test.values
```
